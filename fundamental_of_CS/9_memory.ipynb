## 2. 메모리 계층
<img src="./img/9_1.png" width="40%">


- 메모리 계층 구조에서 위로 올라갈수록 속도가 빨라지고, 용량은 작아짐. 
- 메모리 계층 구조에서 `전달되는 데이터가 아래쪽에 있을 경우 CPU에 도달하려면 위에 있는 모든 계층을 거쳐야함`. (단,`지역성`으로 속도가 느려지지는 않음.)
- 여러 종류의 메모리가 필요한 이유?
    - `속도와 비용 때문` : CPU안의 레지스터가 메모리 중 가장 빠르지만 용량이 가장 작음.  RAM이나 하드디스크보다 비싸기 때문에 적정한 가격과 하드웨어 구현 등 여러 요소를 종합하여 메모리 계층 구조가 완성됨. 

## 3. 지역성과 캐시히트

- 캐시 메모리
    - 레지스터와 메인 메모리 사이에 있음. 
    - CPU가 데이터를 요청하면 메인 메모리에서 해당 데이터와 그 인접 데이터로 이루어진 `메모리 블록`을 캐시로 가져옴. `cache line`이라고 불리우며, 크기는 64~128 바이트정도. 여기서 해당 데이터만 캐시에서 레지스터로 전송
    - CPU가 다른 데이터를 요청하면 우선 캐시에 요청한 데이터가 있는지 확인.
        - 캐시 히트(cache hit) : 요청한 데이터가 캐시에 있는 경우
        - 캐시 미스(cache miss) : 캐시에 없어서 메인 메모리에서 가져와야하는 경우
        
        
- 캐시 메모리가 필요한 이유
    - 메인 메모리에서 데이터를 가져올 때는 20~100사이클의 시간이 걸림. 
    - CPU의 성능이 좋아지면서 데이터를 요청할 때마다 시간이 오래걸리는 문제를 해결하기 위해 캐시 메모리 사용.
    - 캐시에서 데이터를 읽어올 때는 3사이클의 시간이 소요.
    
    
- `지역성 (locality)`
    - `데이터 접근이 같은 메모리 공간이나 인접한 메모리 공간에서 자주 일어난다`는 의미.
    - 시간적 지역성 : 특정 데이터에 한 번 접근했을 때 곧 다시 그 데이터에 접근할 가능성이 높음.
    - 공간적 지역성 : 접근하려는 데이터는 이전에 접근했던 데이터 근처에 있을 확률이 높음.
    - 캐시 히트가 일어나기 위해선 요청한 데이터들이 인접한 메모리 공간에 있어야함.  실제 프로그램이 실행되면서 필요한 데이터가 모여 있을 확률이 매우 높기 때문에 캐시를 충분히 이용해야함

- `파이썬의 리스트` : 메모리 공간에서 연속적으로 있지 않아, 공간적 지역성이 적용되지 않는다.


## 4. 가상 주소 공간
- 프로세스가 생성될 때, 메모리를 할당받음. (32비트 운영체제 시 4GB 할당.)
- 커널 영역 : 운영체제가 담당하는 메모리 영역 `2GB` 씩 받는다
- 유저 영역 : 실제 프로그램이 담당하는 메모리 영역 `2GB` 씩 받음
    - 코드 세그먼트
    - 데이터 세그먼트
    - 스택 세그먼트
    - 힙 세그먼트
    <img src="./img/9_2.png" width="100%">
    (그림에서 화살표는 각 세그먼트가 확장하는 방향을 나타냄. 0x7FFFFFFF 10진수로 2GB를 나타냄) 
    
### 4.1 코드 세그먼트
- `프로그램 인스트럭션`이 저장되는 공간.
- 코드가 인스트럭션으로 변환되어 하드에 저장되어 있다가 프로세스가 실행되면 코드 세그먼트에 올라감.
- 프로세스가 실행 되기 전 크기를 알 수 있음. 

### 4.2 데이터 세그먼트 
- 전역변수가 저장되는 공간
- 전역변수는 프로세스가 실행될 때 데이터 세그먼트에 올라가고, 프로세스가 종료될 때 소멸. 
- 프로세스가 실행 되기 전 크기를 알 수 있음. 

### 4.3 스택 세그먼트
- 지역변수가 저장되는 공간
- 함수를 호출 시 스택 프레임이 스택 세그먼트에 생성됨. 이 때 함수 실행 도중 다른 함수를 호출하면, 호출된 함수의 스택프레임이 호출한 함수의 스택 프레임 위에 쌓이며 확장. 
- 스택 오버플로(Stack over flow) 오류 : 스택 세그먼트가 최대로 늘어날 수 있는 크기를 넘길 때 발생하는 오류

### 4.4 힙 세그먼트
- 프로그래머가 자유롭게 메모리를 할당하고 해제할 수 있는 공간
- 메모리 누수 : 힙 세그먼트에 할당한 메모리는 해제하지 않아 메모리 공간에 계속 남아 있는 상황. 
- 힙은 스택과 달리 늘어날 수 있는 최대 크기가 정해져 있지 않음. 
- 힙 세그먼트의 장단점
    - 장점
        - 메모리의 할당과 해제 시점을 프로그래머가 지정 가능
        - 할당한 메모리를 해제하지 않고 메모리 주소 값을 참존 리턴으로 전달 시 언제나 접근 가능
    - 단점 
        - 메모리 할당 공간을 탐색해야 하기 때문에 느림. 
        - 메모리 단편화 문제 : 할당과 해제가 잦아 메모리의 빈 공간이 나뉘어짐. 관련데이터가 멀리 떨어져 저장될 가능성, 캐시 미스가 높아지게 됩니다. ex) 1, 2, 3, 4 저장후 2를 해제, 100을 등록하여 1, 3, 4, 100 등 이런식으로 데이터가 저장됨
        - `메모리 누수` : 할당된 메모리를 더이상 호출 할 수 없는 경우

- `EIP`(PC) : n, b, a 가 쌓이고난 뒤에, 프로그램 카운터 `PC`가 쌓이고, 다음에 실행할 인스트럭션의 주소 값이 저장됩니다. 함수 호출이 끝난뒤, 이어서 실행할 인스트럭션을 어딘가에 저장해 두어야하기 때문입니다. 

- `EBP`(프레임 포인터) 레지스터
    - 메인스택 프레임의 프레임 포인터 값. 위 그림에서 adder 스택 프레임의 기준이 되게하였음
    - 프레임 포인터는 스택 프레임의 기준
    - 스택프레임에 있는 지역 변수에 접근할 때도 프레임 포인터 이용해 접근


- `스택 포인터` 레지스터
    - esp (extended stack pointer)
    - 스택 세그먼트의 맨 위를 가리키는 스택 포인터 레지스터
    

- `프레임 포인터` 레지스터
    - ebp (extended base pointer) 
    - 프레임 포인터는 `스택 프레임`의 기준
    - 스택프레임에 있는 지역 변수에 접근할 때도 프레임 포인터 이용해 접근
    
    
- `스택 포인터와 프레임 포인터`를 이용해 스택 프레임을 할당하고 `해제`하는 과정
    - `ebp`값을 `esp`에 대입. 실제 메모리를 지우지 않으나 덮어쓰므로 결국 해제한 것과 같음.

## 6. 가상 메모리와 페이징

### 6.1 가상 메모리  
- 가상메모리 (virtural memory)
    - `메인 메모리를 확장`하기 위해 `페이지파일`로 불리는 `하드디스크의 일정 부분을 메인 메모리처럼 사용`하는 것을 말함. 
    - `물리 메모리` : 메인 메모리(RAM) + 하드디스크의 페이지 파일
    - 가상 메모리 관리 기법
       - 세그먼테이션 (segmentation) 기법
       - 페이징 (paging) 기법
       - 오늘날 운영 체제는 대부분 두가지를 함께 사용 (segmentation with paging)
       
 - 가상 주소 공간 
     - 프로세스에 주어지는 메모리 공간
     - 논리 주소 : 가상 주소 공간의 메모리 주소
     - 물리 주소 : 메인 메모리의 주소
     
- MMU
     - 프로세스를 실행하기 위해서는 물리 메모리가 필요함. 논리 주소를 물리 주소로 변환해 메인 메모리를 사용하는데 필요한 하드웨어.

### 6.2 페이징 
- `페이징 기법` : 가상주소 공간 , 메인 메모리를 일정한 크기로 나누어 다루는 것
- 가상 주소 공간 
    - 페이지 : 가상 주소 공간, 을 일정한 크기로 쪼갠 한 부분으로, 
    - VPN : 페이지 순서를 나타내는 비트. 페이지 넘버. 
    - 오프셋 : 페이지 안에서 특정 바이트를 가리키는 비트. 
    - 페이지 넘버와 오프셋을 더하면 논리주소가 되며, 논리주소는 CPU가 요청하는 주소를 의미. 
    
<img src="./img/9_11_1.png" width="40%">


- 메인 메모리
    - 페이지 프레임 : `메인 메모리`를 가상 주소 공간의 페이지와 `같은 크기로 쪼갠` 한 부분으로, `실제로 존재하는 메모리영역`.  
    - PPN : 프레임 넘버, VPN과 1:1대응


- 페이지 테이블
    - 어떤 프로세스의 `VPN, PPN, 상태(유효비트) 등을 저장`하는 테이블.
    - `유효비트`는 프레임이 실제 메인 메모리에 있는지 아니면 하드디스크에 있는지를 나타냄. 
    - 페이지 테이블은 메인 메모리에 저장 
    - 페이지는 페이지 테이블을 통해 프레임으로 대응되어야만 실제 메모리를 사용할 수 있음. 
    - `PTBR` : 페이지 테이블의 시작주소를 가리키는 레지스터

<img src="./img/9_12.png" width="80%">

- 요구 페이징
    - 프로세스를 실행할 때 모든 페이지를 프레임에 매핑하는 것이 아니라, 필요한 페이지만 메인 메모리에 올려 실행하는 것.
    - 매핑한 페이지는 페이지 테이블의 유효비트를 1로 바꿈. 


- 페이지 폴트
    - CPU가 요청한 페이지가 메인 메모리에 없을 때 발생. 페이지 폴트가 발생하면 해당페이지를 하드디스크에서 가져와 빈 프레임에 할당하면 됨.
    - 페이지 폴트가 자주 일어나면 프로그램 성능이 매우 떨어짐.
    - 페이지 테이블상 `valid bit`이 `0`인 경우 발생
    - 유효 비트가 `0`일경우 메인 메모리에 존재하지 않는 경우. 하드디스크에서 가져온다.
    - 페이지 폴트가 일어날 확률을 줄이려면 `지역성을 고려하면서 프로그래밍`을 해야함. 
    

- 변환 색인 버퍼 (TLB)
    - 주소 변환 속도를 높이기 위한 `일종의 캐시`로 최근에 사용한 페이지 테이블 일부 저장.
    - `TLB 히트` 시 메인메모리에 있는 페이지 테이블에 직접 접근하지 않아도 되므로 굉장히 빠름.
